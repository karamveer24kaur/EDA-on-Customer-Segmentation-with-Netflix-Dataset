```python
# Visualize the distribution of age of customers

# make a figure
plt.figure(figsize=(10, 6))

# add the type of plot, histplot categorises diferent age groups, bins define the no. of categories
ax = sns.histplot(df['age'], bins=20, color = 'pink')

# Get the bin heights and edges, to know how many counts of value are in each bin
heights, edges = np.histogram(df['age'], bins=20, )

# Add text annotations for each bin
for height, edge in zip(heights, edges):
    
    # Calculate the position for the text
    x = edge + (edges[1] - edges[0]) / 2  # Center of the bin
    y = height
    
    # Add the text annotation
    ax.text(x, y, str(height), ha='center', va='bottom')

plt.title('Age Distribution of Netflix Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

```python
# insights

# The chart showcases the frequency of Netflix customers across various age groups from 20 to 70 years

# key age groups
# 20s (Age 20-29): Highest frequency with values around 298 to 299 shows  strong engagement among younger audiences
# 30s (Age 30-39): slight decline
# 40s (Age 40-49): further decline
# 50s (Age 50-59): a small rise in frequencies
# 60s (Age 60-69): The highest frequency peaks at 309, suggest srting engagement from older adults
```

```python
# visualize subscription type distribution

plt.figure(figsize=(10, 6))
plot = sns.countplot(data=df, x='subscription_type', palette='viridis')

# Add labels to each bar
for p in plt.gca().patches: # loops through each bar(patch) in the current axes (gca stands for "get current axes")
    plt.annotate(f'{int(p.get_height())}', 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha='center', va='bottom', 
                 fontsize=12)
    
# annotate function : plt.annnotate
# parameters
# 1.) {int(p.get_height())} = gets height of each bar and converts it into an integer
# 2.) (p.get_x() + p.get_width() / 2., = gives x- cordinate for label so that it is in center 
# p.get_height()) = the y cordinate, set to the top of the bar
# 3.) ha='center', va='bottom' = horizontal alignment to center the text and vertical one to put text just above the bar
# 4.) fontsize=12 = sets font size of text

plt.title('Distribution of Subscription Types')
plt.xlabel('Subscription Type')
plt.ylabel('Count')

plt.show()
```

```python
# insights

# subscription counts for all three categories (Basic, Standard, Premium) are relatively similar
# there is no clear dominant subscription type
# customers are equally satisfied with all three offerings
```

```python
# visualize numeric data with correlation plot

# get only numeric data, since correlation can be made between numerical values only
numeric_df = df.select_dtypes(include=['number'])
```

```python
numeric_df
```

```python
# Define groups of features, since a lot of numeric data plotted on correlation plot won't give any actual insight

group1 = ['age', 'watch_hours']
group2 = ['last_login_days', 'monthly_fee']
group3 = ['number_of_profiles', 'avg_watch_time_per_day']
```

```python
# Create a new DataFrame with reordered columns

ordered_columns = group1 + group2 + group3
correlation_matrix = numeric_df[ordered_columns].corr()
```

```python
# Plot the heatmap

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", 
            cbar_kws={'label': 'Correlation Coefficient'}, square=True)

# parameters
# "annot = True" : enables annotation of each cell with the numeric value representing the correlation coefficient
# "Effect" : You will see numerical values displayed inside each cell of the heatmap
# "cmap='coolwarm'" : specifies colormap used
# "fmt=".2f"" : defines the format for displaying the numerical values in the cells
# "cbar_kws={'label': 'Correlation Coefficient'}" : customizes the color bar displayed next to the heatmap
# "square=True" : forces each cell of the heatmap to be square-shaped

plt.title('Correlation Matrix of Numeric Data')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# insights

# interpreting the Correlation Values
# Values close to 1: Strong positive correlation
# Values close to -1: Strong negative correlation
# Values around 0: No correlation

# 1.) age and watch hours, correlation: 0.03, very weak positive correlation
# indicating age has little effect on watch hours

# 2.) watch hours and average watch time per day, correlation: 0.35, moderate positive correlation
# suggesting that as watch hours increase, the average watch time per day also tends to increase

# 3.) last login days and average watch time per day, correlation: -0.34, moderate negative correlation
# indicating that as the number of days since the last login increases, the average watch time per day tends to decrease

# 4.) monthly fee and other variables
# correlations with other variables (including age, watch hours, and average watch time per day) show negligible values
# suggesting that the monthly fee does not significantly correlate with user activity metrics

# 5.) number of profiles and other variables, correlation values (mostly around 0.02) suggest very weak relationships
# indicating the number of profiles has little to no correlation with other metrics.
```
